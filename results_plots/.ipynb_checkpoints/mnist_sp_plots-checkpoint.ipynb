```python
from tensorflow.examples.tutorials.mnist import input_data

import glob
import numpy as np

import seaborn as sns
import matplotlib.pylab as plt
```

### Raw Input

```python
mnist = input_data.read_data_sets("../MNIST")
x_train = mnist.train.images[40:60]
    
def corrupt(X_o,corNum=10):
    X = X_o.copy()
    N,p = X.shape[0],X.shape[1]
    
    for i in range(N):
        loclist = np.random.randint(0, p, size = corNum)
        for j in loclist:
            if X[i,j] > 0.5:
                X[i,j] = 0
            else:
                X[i,j] = 1
    return X

noise_factors = [0.05]
for noise_factor in noise_factors:
    x_train_noisy = corrupt(x_train, corNum = int(noise_factor*784))

    h = w = 28
    # plot of generation
    n = np.sqrt(16).astype(np.int32)
    I_generated = np.empty((h*n, w*n))
    for i in range(n):
        for j in range(n):
            I_generated[i*h:(i+1)*h, j*w:(j+1)*w] = x_train_noisy[i*n+j, :].reshape(28, 28)

    plt.figure(figsize=(4, 4))
    plt.imshow(I_generated, cmap='gray')
    plt.axis('off')
    plt.title("Raw MNIST with 5% Noise")
#     plt.savefig("../sp_pic/raw_sp05")
    plt.show()
```

### Image Table

```python
def plot(x,name,noise):
    h = w = 28
    # plot of generation
    n = np.sqrt(16).astype(np.int32)
    I_generated = np.empty((h*n, w*n))
    for i in range(n):
        for j in range(n):
            I_generated[i*h:(i+1)*h, j*w:(j+1)*w] = x[i*n+j, :].reshape(28, 28)

    plt.figure(figsize=(4, 4))
    plt.imshow(I_generated, cmap='gray')
    plt.axis('off')
    plt.title(name)
#     plt.title("Raw Fashion MNIST with "+str(int(noise_factor*100))+"% Noise")
#     plt.savefig("../sp_pic/rvae_n"+noise+"_l50")
    plt.show()
```

```python
noise = "29"

# RVAE
rvae = np.load("./rvae_sp_noise/lambda_50/noise_0."+noise+"/generation_fid.npy")[22:]
plot(rvae,"rvae",noise)

#VAE
vae =  np.load("./vae_sp_noise/0."+noise+"/generation_fid.npy")
plot(vae,"vae",noise)

#RPCA
rpca = np.load("./rpca_sp_noise/0."+noise+"/generation_fid.npy")
rpca = plot(rpca,"plot",noise)

#LPF
lpf = np.load("./fourier_vae_mnist/0."+noise+"/generation_fid.npy")
lpf = plot(lpf,"lpf",noise)
```

### ADMM 

```python
lambdas = [5,15,25,50,70]
noise_factors2 = [0.33]

losses = {}
for n in noise_factors2:
    tmp = {}
    for l in lambdas:
        path = "rvae_sp_noise/lambda_"+str(l)+"/noise_"+str(n)+"/*_all.npy"
        file_list = glob.glob(path)
        list.sort(file_list, key=lambda x: int(x.split('/')[-1].split('loss')[-1].split('_')[0]))
        loss = []
        for f in file_list:
            loss.append(np.load(f))
        tmp[l] = np.array(loss).reshape(30*20)
    losses[n] = tmp

    
for n in noise_factors:
    for l in lambdas:
        plt.plot(losses[n][l],label=r'$\lambda = $'+str(l))

    plt.legend()
    plt.xlabel("iteration")
    plt.ylabel("loss")
    plt.title("ADMM Training of MNIST with 33% Salt-and-Pepper Noise")
#     plt.savefig("../sp_pic/rvae_loss_admm.png")
    plt.show()
```
